* Определить корни
* Найти интервалы, на которых функция возрастает
* Найти интервалы, на которых функция убывает
* Построить график
* Вычислить вершину
* Определить промежутки, на котором f > 0
* Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
from matplotlib import pyplot as plt
import math
import numpy as np
```

```python
def f(x):
    return -12 * (x**4) * math.sin(math.cos(x)) - 18 * (x**3) + 5*(x**2) + 10*x - 30

x = [t/10 for t in range(-1000, 1000)]
y = [f(t) for t in x]


plt.plot(x, y)
```

Как видно из графика функция имеет бесконечное количество корней. Что подтверждает попытка их нахождения в sympy

```python
x = Symbol('x')
у = -12 * x**4 * sin(cos(x)) - 18 * x**3 + 5*x**2 + 10*x - 30
slv = solve( у, x)
slv
```

Будем рассматривать интервал от -10 до 10

Построим график функции в simpy:

```python
x = symbols('x')
p1 = plot(-12 * (x**4) * sin(cos(x)) - 18 * (x**3) + 5*(x**2) + 10*x - 30, show = False)
p1.show()
```

Из графика: корни на участке от -10 до + 10 примерно равны: -7.6, -5, 4.5 и 8

```python
def f(x):
    return -12 * (x**4) * math.sin(math.cos(x)) - 18 * (x**3) + 5*(x**2) + 10*x - 30

x = [t/100 for t in range(-1000, 1000)]
y = [f(t) for t in x]
plt.axvline(x=-7.6, color="red")
plt.axvline(x=-5, color="red")
plt.axvline(x=4.5, color="red")
plt.axvline(x=8, color="red")
plt.grid()
plt.plot(x, y, color="green")
```

Функция возрастает на участках от -6.9 до -4.2  и от 7 до 10 по оси абсцисс

Функция убывает на участках от -10 до -6.9 и от 4 до 7

```python
plt.axvline(x=-6.9, color="blue")
plt.axvline(x=-4.2, color="blue")
plt.axvline(x=4, color="blue")
plt.axvline(x=7, color="blue")
plt.grid()
plt.plot(x, y, color="green")
```

f > 0 на участках: 
[-10;-7.6], [-5; 4.5], [8;10]
Соответственно f<0 на участках:
[-7.6;-5] [4.5; 8]

Найдем производную: def(-12 * (x**4) * sin(cos(x)) - 18 * (x**3) + 5*(x**2) + 10*x - 30) = -12*4*x**3*sin(cos(x)) + 12* sin(x)* cos(cos(x))*x**4  - 18 * 3* (x**2) + 10*(x) + 10
И построим график

```python
def f(x):
    return -12*4*x**3*sin(cos(x)) + 12* sin(x)* cos(cos(x))*x**4  - 18 * 3* (x**2) + 10*(x) + 10
x = [t/100 for t in range(-1000, 1000)]
y = [f(t) for t in x]
plt.axvline(x=-6.9, color="blue")
plt.axvline(x=-4.2, color="blue")
plt.axvline(x=4, color="blue")
plt.axvline(x=7, color="blue")
plt.grid()
plt.plot(x, y, color="green")
```

Как видно из графика производной корни (переход от максимума в минимум и обратно) найдены верно
